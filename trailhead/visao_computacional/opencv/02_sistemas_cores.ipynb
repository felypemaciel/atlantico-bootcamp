```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def convertcolor(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img = cv2.imread('../img/ponte.png')
cv2.imshow('original', img)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('gray', gray)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('hsv',hsv)

lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2.imshow('lab', lab)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1), plt.imshow(convertcolor(img)), plt.title('original')
plt.subplot(2,2,2), plt.imshow(gray, cmap='gray'), plt.title('GRAY')
plt.subplot(2,2,3), plt.imshow(hsv, cmap='hsv'), plt.title('hsv')
plt.subplot(2,2,4), plt.imshow(lab), plt.title('lab')
```

## Separando os canais da imagem

```python
(canalAzul, canalVerde, canalVermelho) = cv2.split(img)
cv2.imshow('Vermelho', canalVermelho)
cv2.imshow('Verde', canalVerde)
cv2.imshow('Azul', canalAzul)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1), plt.imshow(convertcolor(img)), plt.title('original')
plt.subplot(2,2,2), plt.imshow(canalVermelho, cmap='gray'), plt.title('red')
plt.subplot(2,2,3), plt.imshow(canalVerde, cmap='gray'), plt.title('green')
plt.subplot(2,2,4), plt.imshow(canalAzul, cmap='gray'), plt.title('blue')
```

```python
# juntando os canais novamente
resultado = cv2.merge([canalAzul, canalVerde, canalVermelho])
```

```python
# exibindo nas cores originais
zeros = np.zeros(img.shape[:2], dtype='uint8')
cv2.imshow('red', cv2.merge([zeros, zeros, canalVermelho]))
cv2.imshow('green', cv2.merge([zeros, canalVerde, zeros]))
cv2.imshow('blue', cv2.merge([canalAzul, zeros, zeros]))
cv2.imshow('original', img)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1), plt.imshow(convertcolor(img)), plt.title('original')
plt.subplot(2,2,2), plt.imshow(convertcolor(cv2.merge([zeros, zeros, canalVermelho]))), plt.title('red')
plt.subplot(2,2,3), plt.imshow(cv2.merge([zeros, canalVerde, zeros])), plt.title('green')
plt.subplot(2,2,4), plt.imshow(convertcolor(cv2.merge([canalAzul, zeros, zeros]))), plt.title('blue')
```
