## Histogramas e equalização de imagem

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../img/ponte.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convertendo para preto e branco
cv2.imshow('imagem p&b', img)
# funcao para calcular o histograma da imagem
h = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.figure(figsize=(10,15))
plt.subplot(2,1,1), plt.imshow(img, cmap='gray'), plt.title('imagem p&b')
plt.subplot(2,1,2)
plt.title('histograma p&b')
plt.xlabel('intensidade')
plt.ylabel('quantidad de pixels')
plt.plot(h)
plt.xlim([0, 256])
plt.show()
cv2.waitKey(0)
```

```python
# outra forma de plotar o histograma
plt.hist(img.ravel(), 256, [0, 256])
plt.show()
```

### Histograma da imagem colorida

**zip()** cria uma lista de tuplas formada pela união das listas passadas e não tem a ver com compactação. 

```python
img = cv2.imread('../img/ponte.png')
cv2.imshow('imagem colorida', img)
# separando os canais
canais = cv2.split(img)
cores = ('b', 'g', 'r')
plt.figure(figsize=(10,5))
plt.title('histograma colorido')
plt.xlabel('intensidade')
plt.ylabel('número de pixels')
for (canal, cor) in zip(canais, cores):
    hist = cv2.calcHist([canal], [0], None, [256], [0, 256])
    plt.plot(hist, color=cor)
plt.xlim([0, 256])
plt.show()

cv2.waitKey(0)
```

### Equalização de Histograma

Para aumenar o constraste da iamgem. Distribui de forma mais uniforme as intensidades dos pixels sobre a imagem.

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
heq = cv2.equalizeHist(img)

plt.figure(figsize=(20, 7.5))

plt.subplot(2,2,1)
plt.title('histograma equalizado')
plt.ylabel('qtde de pixels')
plt.hist(heq.ravel(), 256, [0,256])
plt.xlim([0, 256])

plt.subplot(2,2,2)
plt.imshow(heq, cmap='gray')
plt.title('imagem equalizada')

plt.subplot(2,2,3)
plt.title('histograma original')
plt.ylabel('qtde de pixels')
plt.hist(img.ravel(), 256, [0,256])
plt.xlim([0, 256])

plt.subplot(2,2,4)
plt.imshow(img, cmap='gray')
plt.title('imagem original')

plt.show()
```
