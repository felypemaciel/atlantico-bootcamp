```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cvtrgb(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Suavização por média

**vstack()** e **hstack()** juntam as imagens numa única imagem final.

```python
img = cv2.imread('../img/ponte.png')
img = img[::2,::2] # diminui a imagem
suave = np.vstack([
    np.hstack([img, cv2.blur(img, (3,3))]),
    np.hstack([cv2.blur(img, (5,5)), cv2.blur(img, (7,7))]),
    np.hstack([cv2.blur(img, (9,9)), cv2.blur(img, (11,11))])
])
```

```python
cv2.imshow("imagens suavizadas (blur)", suave)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(13, 12))
plt.subplot(3,2,1), plt.imshow(cvtrgb(img))
plt.subplot(3,2,2), plt.imshow(cvtrgb(cv2.blur(img, (3,3))))
plt.subplot(3,2,3), plt.imshow(cvtrgb(cv2.blur(img, (5,5))))
plt.subplot(3,2,4), plt.imshow(cvtrgb(cv2.blur(img, (7,7))))
plt.subplot(3,2,5), plt.imshow(cvtrgb(cv2.blur(img, (9,9))))
plt.subplot(3,2,6), plt.imshow(cvtrgb(cv2.blur(img, (11,11))))
```

## Suavização por mediana

```python
suave = np.vstack([
    np.hstack([img, cv2.medianBlur(img, 3)]),
    np.hstack([cv2.medianBlur(img, 5), cv2.medianBlur(img, 7)]),
    np.hstack([cv2.medianBlur(img, 9), cv2.medianBlur(img, 11)])
])
cv2.imshow('mediana', suave)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(13, 12))
plt.subplot(3,2,1), plt.imshow(cvtrgb(img))
plt.subplot(3,2,2), plt.imshow(cvtrgb(cv2.medianBlur(img, 3)))
plt.subplot(3,2,3), plt.imshow(cvtrgb(cv2.medianBlur(img, 5)))
plt.subplot(3,2,4), plt.imshow(cvtrgb(cv2.medianBlur(img, 7)))
plt.subplot(3,2,5), plt.imshow(cvtrgb(cv2.medianBlur(img, 9)))
plt.subplot(3,2,6), plt.imshow(cvtrgb(cv2.medianBlur(img, 11)))

plt.show()
```

## Suavização com filtro bilateral

Mais lento, porém preserva as bordas e o ruído é removido. É utilizado um filtro gaussiano do espaço ao redor do pixel e outro guassiano que leva em conta a diferença de intensidade entre os pixels.

```python
suave = np.vstack([
    np.hstack([img, cv2.bilateralFilter(img, 3, 21, 21)]),
    np.hstack([cv2.bilateralFilter(img, 5, 35, 35), cv2.bilateralFilter(img, 7, 49, 49)]),
    np.hstack([cv2.bilateralFilter(img, 9, 63, 63), cv2.bilateralFilter(img, 11, 77, 77)])
])

cv2.imshow('filtro bilateral', suave)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(13, 12))
plt.subplot(3,2,1), plt.imshow(cvtrgb(img))
plt.subplot(3,2,2), plt.imshow(cvtrgb(cv2.bilateralFilter(img, 3, 21, 21)))
plt.subplot(3,2,3), plt.imshow(cvtrgb(cv2.bilateralFilter(img, 5, 35, 35)))
plt.subplot(3,2,4), plt.imshow(cvtrgb(cv2.bilateralFilter(img, 7, 49, 49)))
plt.subplot(3,2,5), plt.imshow(cvtrgb(cv2.bilateralFilter(img, 9, 63, 63)))
plt.subplot(3,2,6), plt.imshow(cvtrgb(cv2.bilateralFilter(img, 11, 77, 77)))

plt.show()
```
